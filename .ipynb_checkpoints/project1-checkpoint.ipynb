```python
%matplotlib inline

from modsim import *
from pandas import read_html

#setup male data tables
male_data_source = 'Data/MaleWhaleData.html'
male_data = read_html(male_data_source, header=1, index_col=0)
male_table = male_data[0]
male_table.columns = ['name', 'born', 'died']

#setup female data tables
female_data_source = 'Data/FemaleWhaleData.html'
female_data = read_html(female_data_source, header=1, index_col=0)
female_table = female_data[0]
female_table.columns = ['name', 'born', 'died', 'offspring']
```

```python
system = System(
    t_start = 1915,
    t_end = 2017
)

male_birth_year = male_table.born
female_birth_year = female_table.born
male_death_year = male_table.died
female_death_year = female_table.died

male_table_size = 85
female_table_size = 74




def determine_living_male_whales(current_year):
    current_male_whales = 0
    for i in linrange(2, male_table_size):
        if male_birth_year[i] <= current_year:
            if male_death_year[i] >= current_year or male_death_year[i] == 0:
                current_male_whales += 1
    return current_male_whales

def determine_living_female_whales(current_year):
    current_female_whales = 0
    for i in linrange(2, female_table_size):
        if female_birth_year[i] <= current_year:
            if female_death_year[i] >= current_year or female_death_year[i] == 0:
                current_female_whales += 1
    return current_female_whales

def count_living_whales(t_start, t_end):
    results = TimeSeries()
    for t in linrange(t_start, t_end):
            results[t] = determine_living_male_whales(t) + determine_living_female_whales(t)
    return results


raw_pop_data = count_living_whales(system.t_start, system.t_end)
newfig()
plot(raw_pop_data, label = 'Raw Data')
decorate(xlabel='Year', 
         ylabel='Number of Living Whales',
         title='Southern Resident Killer Whale Population Data')
```
